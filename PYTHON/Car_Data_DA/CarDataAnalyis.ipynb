```python
import pandas as pd
data = pd.read_csv("2. Cars Data1 (1).csv")
data
```

```python
datacopy = data.copy()
datacopy.head()
```

```python
datacopy.isnull().sum()
```

```python
#For removing empty rows in the csv file
datacopy = datacopy.dropna(how='all')
datacopy.isnull().sum()
```

```python
#For filling isnan with mean value of the column
datacopy['Cylinders'].fillna(datacopy['Cylinders'].mean(),inplace = True)
datacopy.isnull().sum()
```

```python
CarMake = datacopy['Make'].value_counts()
CarMake
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
datacopy['Make'].value_counts().nlargest(5).plot(kind='bar', color='green')
plt.title('Number of Cars by Make')
plt.xlabel('Make')
plt.ylabel('Number of Cars')
plt.show()
```

```python
datacopy['Origin'].unique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(3,5))
datacopy['Origin'].value_counts().plot(kind='bar', color='red')
plt.title('Number of Cars by Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Cars')
plt.show()
```

```python
datacopy['Price'] = datacopy['Invoice'].str.replace('$',"")
datacopy.head()
```

```python
datacopy.shape
```

```python
datacopy['Price'] = datacopy['Price'].str.replace(',','')
datacopy['Price'] = datacopy['Price'].astype(int)
datacopy['Price'].dtype
```

```python
datacopy.head()
```

```python
average_price = datacopy['Price'].mean()
average_price
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average price for each make
average_price_by_make = datacopy.groupby('Make')['Price'].mean()

# Select the top 5 makes with the highest average prices
top_5_makes = average_price_by_make.nlargest(5).index

# Filter the data for the top 5 makes
top_5_data = datacopy[datacopy['Make'].isin(top_5_makes)]

# Create a bar plot for the top 5 makes
plt.figure(figsize=(12, 8))
sns.barplot(x='Make', y='Price', data=top_5_data, hue='Origin', palette='muted')

# Set labels and title
plt.title('Top 5 Car Makes by Average Price', fontsize=20)
plt.xlabel('Make', fontsize=10)
plt.ylabel('Price', fontsize=10)
plt.legend(loc='upper left', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
datacopy = datacopy.sort_values(by='Price',ascending=False)
datacopy_15 = datacopy.head(15)

# Create a bar plot
sns.barplot(x = 'Make', y = 'Price', data = datacopy_15, hue = 'Origin',palette='Set1')


# Set labels and titles
plt.title("Price Distribution of Top 7 Car Types by Make and Origin", fontsize=20)
plt.xlabel("Make",fontsize=14)
plt.ylabel("Price",fontsize=14)
plt.xticks(rotation=90)
plt.legend(loc='upper left',fontsize=14)
plt.show()
```

```python
sns.set_style("whitegrid")
# create a bar plot with larger size
plt.figure(figsize=(13,7))
sns.barplot(x = 'Make', y = 'Price', data = datacopy_15, hue = 'Origin', palette='Blues')

#set labels and titles
plt.title("Price Distribution by No. of Make", fontsize=16)
plt.xlabel("Make",fontsize=8)
plt.ylabel("Price",fontsize=8)
plt.show()
```

```python
datacopy_grouped = datacopy.groupby('Make').mean()[['Cylinders']]
datacopy_grouped.sort_values(by='Cylinders',ascending=False,inplace=True)
top_10_car_types = datacopy_grouped.head(10)
top_10_car_types
```

```python
plt.figure(figsize=(12,8))
#create a bar plot of the average cylinder size by top 10 car types
sns.barplot(x=top_10_car_types.index, y='Cylinders', palette='PuBu', data=top_10_car_types)

#set the title and labels
plt.title("Average Cylinder Size by Top 10 Car Types", fontsize=16)
plt.xlabel("Car Type")
plt.ylabel("Average Cylinder Size")
plt.xticks(rotation=90) # to rotate x_axis label by 90 degrees
plt.show()
```

```python
# Create a bar plot
sns.lmplot(x = 'EngineSize', y = 'Horsepower', data = datacopy, hue = 'Origin',palette='Set1')


# Set labels and titles
plt.title("Relationship between EngineSize and HorsePower", fontsize=20)
plt.xlabel("Engine Size",fontsize=14)
plt.ylabel("Horsepower",fontsize=14)
plt.xticks(rotation=90)
plt.legend(loc='upper left',fontsize=14)
plt.show()
```
